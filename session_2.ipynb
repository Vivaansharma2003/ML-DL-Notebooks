## Day 2: Classification Algorithms & SVM


### 1. Objectives:

In the previous session, we predicted numbers (Regression). Now, we will predict Categories (Classification).

Scenario: Based on sensor data (Temperature & Vibration), is the machine "Normal" (0) or "Failing" (1)?

Logistic Regression: Drawing a line based on probability.

SVM: Drawing the widest possible road (Margin) between classes.
